## Introduction to Latent Semantic Analysis

In 1975, G. Salton et al. proposed Vector Space Model (VSM) in their paper ["A Vector Space Model for Automatic Indexing"][1].This one paper makes the assumption that texts are made up of words, regardless of their order, grammar, or meaning.

On the basis of this paper, the concept of a **Bag of Words Model** was proposed.

The basic idea of VSM: map a text into word vectors, and multiple texts form a word vector space.

- Step 1: Create a sequential word list
- Step 2: Calculate the word TF-IDF value of the article.
- Step 3: Corresponding the TF-IDF value of a word to its position in the vocabulary yields a vector.
- Step 4: Get the word vector space

[1]: https://dl.acm.org/doi/10.1145/361219.361220

：每一列代表一个文档（例如一封邮件）。如果您有10000封邮件，矩阵就有10000列。
行：每一行代表一个独特的词项（如单词或短语）。行的总数是您的文档集（即所有邮件）中所有独特词项的数量。
元素值：在TF-IDF矩阵中，每个元素的值是根据特定词项在特定文档中的重要性来计算的，这通常由两部分组成：
TF（Term Frequency）：词频，即一个词在特定文档中出现的次数相对于该文档中所有词的总数的比例。
IDF（Inverse Document Frequency）：逆文档频率，这是一个度量，反映了一个词在整个文档集中的常见程度。IDF值越高，表示该词越有可能具有区分力。

The principle of VSM is simple and intuitive, but there are obvious drawbacks：

1. For text with a relatively large amount of data, the calculation speed is slow.
2. For scenarios with a high number of synonyms and near-synonyms, the results obtained by VSM are not satisfactory.
3. The VSM assumes that words are completely independent of each other, but in practice this doesn't follow our usual logic.

<font size='5.5'><b>So our character LSA - Latent Semantic Analysis</b></font>

Based on VSM, LSA uses Truncated Singular Value Decomposition (TSVD), a dimensionality reduction technique, to the lexical-document matrix in an attempt to expose the implicit semantic relationships that exist between words and documents.


### SVD - Singular Value Decomposition

Given a matrix $A$, the Singular Value Decomposition (SVD) is a factorization of the form:

$$
A = U\Sigma V^T
$$

Where:
- $U$ is an $m \times m$ orthogonal matrix, whose columns are the left singular vectors of $A$.
- $\Sigma$ is an $m \times n$ diagonal matrix, with its diagonal elements being the singular values of $A$. The non-diagonal elements of $\Sigma$ are zero.
- $V$ is an $n \times n$ orthogonal matrix, whose columns are the right singular vectors of $A$.
- $V^T$ is the transpose of $V$.

The singular values in $\Sigma$ are non-negative and are usually arranged in descending order. These singular values represent the magnitude of the principal components of the matrix $A$.


<img height = "300" src="">

For **singular values**, they are similar to **eigenvalues** in our eigenvalue decomposition, and in the singular value matrix, they are also arranged from largest to smallest. Moreover, the decrease of singular values is particularly rapid. In many cases, the sum of the top 10% or even 1% of singular values accounts for more than 99% of the total sum of singular values. This means that we can also use the largest *k* singular values and their corresponding left and right singular vectors to approximate the description of the matrix. In other words:

$$ A_{m \times n} = U_{m \times m}E_{m \times n} \approx U_{n \times k} \Sigma_{k \times k} V_{n \times k}^T $$

Where *k* is much smaller than *n*, meaning a large matrix *A* can be represented by three smaller matrices $U_{m \times k}, \Sigma_{k \times k}, V_{n \times k}^T$.


对于**奇异值**，它跟我们特征分解中的**特征值**类似，在奇异值矩阵中也是按照从大到小排列。而且奇异值的减少特别的快，在很多情况下，前10％甚至1%的奇异值的和就占了全部的奇异值之和的99％以上的比例。也就是说，我们也可以用最大的*k*个的奇异值和对应的左右奇异向量来近似描述矩阵。也就是说：
$$ A_{m \times n} = U_{m \times m}E_{m \times n} \approx U_{n \times k} \Sigma_{k \times k} V_{n \times k}^T $$

其中*k*要比*n*小很多，也就是一个大的矩阵*A*可以用三个小的矩阵 $U_{m \times k}, \Sigma_{k \times k}, V_{n \times k}^T$来表示。


在潜在语义分析（LSA）中，截断奇异值分解（Truncated Singular Value Decomposition, TSVD）是一个关键步骤。TSVD 是奇异值分解（SVD）的一种变体，用于降维和提取文本数据的重要特征。给定一个词项-文档矩阵 $A$，其行代表不同的词项，列代表不同的文档，TSVD 将这个矩阵分解为三个矩阵：$U$、$\Sigma$ 和 $V^T$。

- $U$ 是一个 $m \times k$ 的矩阵，包含了 $A$ 的左奇异向量。
- $\Sigma$ 是一个 $k \times k$ 的对角矩阵，其对角线上的元素是 $A$ 的前 $k$ 个最大的奇异值。
- $V^T$ 是一个 $k \times n$ 的矩阵，包含了 $A$ 的右奇异向量。

在 TSVD 中，选择 $k$（一个比原始矩阵列数少的数）是一个关键决策，它决定了保留的奇异值的数量。通过保留最大的 $k$ 个奇异值，TSVD 可以去除噪声和不重要的信息，同时保留数据中最重要的结构和模式。这种方法在处理大规模文本数据时特别有效，因为它可以显著减少数据的维度，同时保留关键信息。


### TSVD - Truncated Singular Value Decomposition

In Latent Semantic Analysis (LSA), Truncated Singular Value Decomposition (TSVD) is a crucial step. TSVD is a variant of Singular Value Decomposition (SVD) used for dimensionality reduction and extracting significant features from text data. Given a term-document matrix $A$, with rows representing different terms and columns representing different documents, TSVD decomposes this matrix into three matrices: $U$, $\Sigma$, and $V^T$.

- $U$ is an $m \times k$ matrix, containing the left singular vectors of $A$.
- $\Sigma$ is a $k \times k$ diagonal matrix, whose diagonal elements are the top $k$ largest singular values of $A$.
- $V^T$ is a $k \times n$ matrix, containing the right singular vectors of $A$.


<img height = "300" src="">

In TSVD, choosing $k$—a number smaller than the original number of columns of the matrix—is a key decision, as it determines the number of singular values retained. By keeping only the largest $k$ singular values, TSVD can eliminate noise and less important information, while preserving the most significant structures and patterns in the data. This approach is particularly effective in handling large-scale text data, as it significantly reduces the dimensions of the data while retaining key information.

Typically, when we are writing a piece of text, our logic is roughly as follows: first we have the topic (theme) of the article, then we write it, and finally we form the article. This is, in part, what LSA and TSVD is all about.

<b><font size="5.5">Topic -> Word -> Text</b></font>

Now, suppose we have a text considered as our input dataset, after the SVD decomposition：

- $A$ is the frequency matrix of how often each word occurs within this piece of text
- $U$ is the topic matrix of the text, where each column can be considered as a topic
- $\Sigma V^T$ it's a "topic-text" matrix.

The advantage of this is that we can avoid problems caused by, for example, near-synonyms and synonyms by gaining a deeper understanding of the relationships between words and sentences and the whole text.

For example, if we come across the word "bank", we may think it means (money) bank if it is in the context of finance, but if it is in the context of fishing, etc., we may think it means river bank.

## Longer text application

Now let's test the LSA in some two larger text. The first one is a fairy tale about a little girl in the magic forest. and the second one is a zombie story about the end of world is coming.


```python
# read the two txt files in to dataframes, and keep the file name as a column
with open("The Enchanted Forest.txt", "r") as f:
    enchanted_forest = f.readlines()

with open("The Last Stand.txt", "r") as f:
    last_stand = f.readlines()

# create a dataframe with the two texts
# one column names text, include the text from the book, one column named tale
# set the column tale, if the text is from the enchanted forest, set the value to True, otherwise false

df = pd.DataFrame(
    {
        "text": enchanted_forest + last_stand,
        "tale": [True] * len(enchanted_forest) + [False] * len(last_stand),
    }
)


# remove the \n from the end of each line then remove the empty lines
df["text"] = df["text"].str.strip()
df = df[df["text"] != ""]

# shuffle the dataframe
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=1, stop_words="english")
bag_of_words = vectorizer.fit_transform(df["text"])
```

```python
from sklearn.decomposition import TruncatedSVD

tsvd = TruncatedSVD(
    n_components=2, random_state=42
)  # keep 2 components because we have 2 sources
lsa = tsvd.fit_transform(bag_of_words)
lsa
```

```python
# combian with the origin df
df1 = df.copy()
df1["topic_1"] = lsa[:, 0]
df1["topic_2"] = lsa[:, 1]
df1["abs_topic_1"] = np.abs(df1["topic_1"])
df1["abs_topic_2"] = np.abs(df1["topic_2"])

df.head(10)
```

```python
# plot two scatter plots, one is stop words removed, one is not
plt.figure(figsize=(12, 6))

# remove the stop words
plt.subplot(1, 2, 1)
plt.title("Stop words removed")
sns.scatterplot(data=df1, x="abs_topic_1", y="abs_topic_2", hue="tale")


# No remove the stop words
df2 = df.copy()
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(df2["text"])

tsvd = TruncatedSVD(n_components=2, random_state=42)

lsa = tsvd.fit_transform(bag_of_words)

df2["topic_1"] = lsa[:, 0]
df2["topic_2"] = lsa[:, 1]
df2["abs_topic_1"] = np.abs(df2["topic_1"])
df2["abs_topic_2"] = np.abs(df2["topic_2"])

plt.subplot(1, 2, 2)
plt.title("Stop words not removed")
sns.scatterplot(data=df2, x="abs_topic_1", y="abs_topic_2", hue="tale")
```

```python
# visualization for the top words in the topic

vectorizer = CountVectorizer(stop_words="english")
bag_of_words = vectorizer.fit_transform(df["text"])

tsvd = TruncatedSVD(n_components=2, random_state=42)
lsa = tsvd.fit_transform(bag_of_words)

dictionary_df = pd.DataFrame(
    tsvd.components_.T,
    index=vectorizer.get_feature_names_out(),  # From scikit-learn 0.22 we have to use get_feature_names_out() instead of get_feature_names()
    columns=["topic_1", "topic_2"],
)

dictionary_df["abs_topic_1"] = np.abs(dictionary_df["topic_1"])
dictionary_df["abs_topic_2"] = np.abs(dictionary_df["topic_2"])

dictionary_df
```

## Implementting LSA manually

#### Manual Latent Semantic Analysis (LSA) Logic
1. Step 1: Text Preprocessing and Tokenization
    Process the cleaned texts and tokenize each document into a set of words.

2. Step 2: Creating Vocabulary
    Create a set of unique vocabulary from the words present in all documents.

3. Step 3: Building the Term-Document Matrix
- If TF-IDF is not used (tfidf=False), compute the frequency of each word in each document directly.
- If TF-IDF is used (tfidf=True), calculate the Term Frequency (TF) and Inverse Document Frequency (IDF) for each word, and    then multiply them to get the TF-IDF matrix.

4. Step 4: Performing Singular Value Decomposition (SVD)
- If manual SVD is used (use_manual_svd=True), use a custom SVD function to decompose the term-document matrix.
- If manual SVD is not used (use_manual_svd=False), use standard library functions (e.g., numpy.linalg.svd) for decomposition.

5. Step 5: Conducting LSA Processing
Extract the first n_topics topics from the singular value decomposition. This involves selecting the first n_topics columns from the U matrix and the first n_topics singular values from the S matrix, and then calculating their product to obtain the LSA matrix.

6. Step 6: Returning Results
Output the shape and content of the LSA matrix and return it as the function's result.

```python
class ManualLSA:
    """
    Manual Latent Semantic Analysis (LSA) Class writen by group 5

    - Attention: This class is not optimized for speed and memory usage!!!
    - Attention: This class supports only English and French stop words!!!
    - Attention: This class suppose that the data is already cleaned!!!

    This class implements a manual version of Latent Semantic Analysis (LSA) for text data.
    It allows for the option of using either standard SVD (Singular Value Decomposition) or a manual implementation of SVD.
    The class provides functionality to preprocess text data, construct a term-document matrix with or without TF-IDF weighting,
    perform SVD, and compute the LSA matrix.

    Key functionalities include:
    - Tokenization of text data and creation of a vocabulary set.
    - Building a term-document matrix, with an option to apply TF-IDF.
    - Performing Singular Value Decomposition (SVD), with a choice between a manual implementation and numpy's SVD function.
    - Storing the U, Sigma, and V^T matrices internally for later use.
    - Fitting the model on text data and transforming the data into its LSA representation.

    Example usage:
        lsa = ManualLSA(tfidf=True, n_topics=3, use_manual_svd=True)
        lsa.fit(cleaned_texts)  # Fit the model on the cleaned text data
        lsa_matrix = lsa.transform()  # Get the LSA representation of the data
    """

    def __init__(
        self,
        tfidf=True,
        n_topics=5,
        stop_words="english",
        max_features=1000,
        use_manual_svd=False,
        use_df_threshold=True,
        min_df_threshold=0.05,
    ):
        self.tfidf = tfidf
        self.n_topics = n_topics
        self.stop_words = stop_words
        self.max_features = max_features
        self.use_manual_svd = use_manual_svd
        self.U = None
        self.Sigma = None
        self.VT = None
        self.LSA = None
        self.use_df_threshold = use_df_threshold
        self.min_df_threshold = min_df_threshold

    def get_stop_words(self):
        # Return the stop words
        current_dir = os.getcwd()
        import json

        path = os.path.join(current_dir, "stop_words.json")

        with open(path) as f:
            stop_words = json.load(f)
        if self.stop_words == "english" or self.stop_words == "English":
            self.stop_words = stop_words["english"]
        elif self.stop_words == "french" or self.stop_words == "French":
            self.stop_words = stop_words["french"]
        else:
            self.stop_words = []

    def to_lower(self):
        # Convert the texts to lower case
        self.cleaned_texts = [text.lower() for text in self.cleaned_texts]

    def tokenize(self):
        # Tokenize the cleaned texts
        self.tokenized_texts = [
            [word for word in text.split() if word.lower() not in self.stop_words]
            for text in self.cleaned_texts
        ]

    def create_vocabulary(self):
        from collections import Counter

        # Count the frequency and document frequency of each word
        total_word_freq = Counter()
        self.document_freq = Counter()

        for text in self.tokenized_texts:
            total_word_freq.update(text)
            unique_words = set(text)
            self.document_freq.update(unique_words)

        # Set thresholds
        min_freq = 2  # Minimum frequency to include a word
        min_df = (
            len(self.tokenized_texts) * self.min_df_threshold
        )  # Minimum document frequency

        # Filter vocabulary based on frequency and document frequency
        self.vocabulary = [
            word
            for word, freq in total_word_freq.items()
            if freq >= min_freq and self.document_freq[word] >= min_df
        ]

    def build_term_document_matrix(self):
        # Initialize term-document matrix
        self.term_document_matrix = np.zeros(
            (len(self.vocabulary), len(self.tokenized_texts))
        )

        # Fill the term-document matrix
        for i, text in enumerate(self.tokenized_texts):
            for word in text:
                if word in self.vocabulary:
                    idx = self.vocabulary.index(word)
                    self.term_document_matrix[idx, i] += 1

        # If TF-IDF is used, compute TF-IDF weights
        if self.tfidf:
            df = np.array([self.document_freq[word] for word in self.vocabulary])
            idf = np.log(len(self.tokenized_texts) / (df + 1))
            idf_matrix = idf[:, np.newaxis]
            self.term_document_matrix *= idf_matrix

        elif self.tfidf:
            # Compute TF-IDF
            tf = np.zeros((len(self.vocabulary), len(self.tokenized_texts)))
            df = np.zeros(len(self.vocabulary))
            for i, text in enumerate(self.tokenized_texts):
                for word in text:
                    if word in self.vocabulary:
                        tf[self.vocabulary.index(word), i] += 1
                for word in set(text):
                    if word in self.vocabulary:
                        df[self.vocabulary.index(word)] += 1
            idf = np.log(len(self.tokenized_texts) / (df + 1))
            idf_matrix = idf[:, np.newaxis]
            self.term_document_matrix = tf * idf_matrix

        else:
            # Compute TF
            tf = np.zeros((len(self.vocabulary), len(self.tokenized_texts)))
            for i, text in enumerate(self.tokenized_texts):
                for word in text:
                    if word in self.vocabulary:
                        tf[self.vocabulary.index(word), i] += 1
            self.term_document_matrix = tf

    def manual_tsvd(self, A):
        # Compute A^TA and AA^T
        AtA = np.dot(A.T, A)
        AAt = np.dot(A, A.T)

        # Compute the eigenvalues and eigenvectors of A^TA and AA^T
        eigvals_ATA, eigvecs_ATA = np.linalg.eig(AtA)
        eigvals_AAT, eigvecs_AAT = np.linalg.eig(AAt)

        # Sort the eigenvalues and eigenvectors
        idx = eigvals_ATA.argsort()[::-1]
        eigvals_ATA = eigvals_ATA[idx]
        eigvecs_ATA = eigvecs_ATA[:, idx]

        # Compute Sigma
        Sigma = np.sqrt(np.diag(eigvals_ATA))

        # Truncate Sigma, U, and V
        Sigma = Sigma[: self.n_topics, : self.n_topics]
        U = eigvecs_AAT[:, : self.n_topics]
        V = eigvecs_ATA[:, : self.n_topics]

        return U, Sigma, V.T

    def auto_svd(self):
        if not self.use_manual_svd:
            self.U, s, self.VT = np.linalg.svd(
                self.term_document_matrix.T, full_matrices=False
            )
            # Truncate U, Sigma, and VT
            self.U = self.U[:, : self.n_topics]
            self.Sigma = np.diag(s)[: self.n_topics, : self.n_topics]
            self.VT = self.VT[: self.n_topics, :]
        else:
            self.U, self.Sigma, self.VT = self.manual_tsvd(self.term_document_matrix.T)

        # Compute LSA
        self.LSA = self.VT[: self.n_topics, :]

    def fit(self, cleaned_texts):
        self.cleaned_texts = cleaned_texts
        self.to_lower()
        self.get_stop_words()
        self.tokenize()
        self.create_vocabulary()
        self.build_term_document_matrix()
        self.auto_svd()

    def fit_transform(self, cleaned_texts):
        # Fit the model and return the LSA matrix
        self.fit(cleaned_texts)
        return self.LSA

    def get_topics(self, n_words=3):
        # Return the top n_words for each topic
        topics = []
        for i, comp in enumerate(self.LSA):
            terms_comp = zip(self.vocabulary, comp)
            sorted_terms = sorted(terms_comp, key=lambda x: x[1], reverse=True)[
                :n_words
            ]
            topics.append(sorted_terms)
        return topics

    def get_top_keywords(self, n_words=10):
        # Return the top n_words keywords
        from collections import Counter

        keyword_freq = Counter()

        for i, comp in enumerate(self.LSA):
            terms_comp = zip(self.vocabulary, comp)
            sorted_terms = sorted(terms_comp, key=lambda x: x[1], reverse=True)
            for term, _ in sorted_terms:
                keyword_freq[term] += 1

        top_keywords = [word for word, freq in keyword_freq.most_common(n_words)]
        return top_keywords
```
