## Introduction to Latent Semantic Analysis

In 1975, G. Salton et al. proposed Vector Space Model (VSM) in their paper ["A Vector Space Model for Automatic Indexing"][1].This one paper makes the assumption that texts are made up of words, regardless of their order, grammar, or meaning.

On the basis of this paper, the concept of a **Bag of Words Model** was proposed.

The basic idea of VSM: map a text into word vectors, and multiple texts form a word vector space.

- Step 1: Create a sequential word list
- Step 2: Calculate the word TF-IDF value of the article.
- Step 3: Corresponding the TF-IDF value of a word to its position in the vocabulary yields a vector.
- Step 4: Get the word vector space

[1]: https://dl.acm.org/doi/10.1145/361219.361220

The principle of VSM is simple and intuitive, but there are obvious drawbacks：

1. For text with a relatively large amount of data, the calculation speed is slow.
2. For scenarios with a high number of synonyms and near-synonyms, the results obtained by VSM are not satisfactory.
3. The VSM assumes that words are completely independent of each other, but in practice this doesn't follow our usual logic.

<font size='5.5'><b>So our character LSA - Latent Semantic Analysis</b></font>

Based on VSM, LSA uses Truncated Singular Value Decomposition (TSVD), a dimensionality reduction technique, to the lexical-document matrix in an attempt to expose the implicit semantic relationships that exist between words and documents.


### SVD - Singular Value Decomposition

Given a matrix $A$, the Singular Value Decomposition (SVD) is a factorization of the form:

$$
A = U\Sigma V^T
$$

Where:
- $U$ is an $m \times m$ orthogonal matrix, whose columns are the left singular vectors of $A$.
- $\Sigma$ is an $m \times n$ diagonal matrix, with its diagonal elements being the singular values of $A$. The non-diagonal elements of $\Sigma$ are zero.
- $V$ is an $n \times n$ orthogonal matrix, whose columns are the right singular vectors of $A$.
- $V^T$ is the transpose of $V$.

The singular values in $\Sigma$ are non-negative and are usually arranged in descending order. These singular values represent the magnitude of the principal components of the matrix $A$.


<img height = "300" src="">

For **singular values**, they are similar to **eigenvalues** in our eigenvalue decomposition, and in the singular value matrix, they are also arranged from largest to smallest. Moreover, the decrease of singular values is particularly rapid. In many cases, the sum of the top 10% or even 1% of singular values accounts for more than 99% of the total sum of singular values. This means that we can also use the largest *k* singular values and their corresponding left and right singular vectors to approximate the description of the matrix. In other words:

$$ A_{m \times n} = U_{m \times m}E_{m \times n} \approx U_{n \times k} \Sigma_{k \times k} V_{n \times k}^T $$

Where *k* is much smaller than *n*, meaning a large matrix *A* can be represented by three smaller matrices $U_{m \times k}, \Sigma_{k \times k}, V_{n \times k}^T$.

### TSVD - Truncated Singular Value Decomposition

In Latent Semantic Analysis (LSA), Truncated Singular Value Decomposition (TSVD) is a crucial step. TSVD is a variant of Singular Value Decomposition (SVD) used for dimensionality reduction and extracting significant features from text data. Given a term-document matrix $A$, with rows representing different terms and columns representing different documents, TSVD decomposes this matrix into three matrices: $U$, $\Sigma$, and $V^T$.

- $U$ is an $m \times k$ matrix, containing the left singular vectors of $A$.
- $\Sigma$ is a $k \times k$ diagonal matrix, whose diagonal elements are the top $k$ largest singular values of $A$.
- $V^T$ is a $k \times n$ matrix, containing the right singular vectors of $A$.


<img height = "300" src="">

In TSVD, choosing $k$—a number smaller than the original number of columns of the matrix—is a key decision, as it determines the number of singular values retained. By keeping only the largest $k$ singular values, TSVD can eliminate noise and less important information, while preserving the most significant structures and patterns in the data. This approach is particularly effective in handling large-scale text data, as it significantly reduces the dimensions of the data while retaining key information.

Typically, when we are writing a piece of text, our logic is roughly as follows: first we have the topic (theme) of the article, then we write it, and finally we form the article. This is, in part, what LSA and TSVD is all about.

<b><font size="5.5">Topic -> Word -> Text</b></font>


The advantage of this is that we can avoid problems caused by, for example, near-synonyms and synonyms by gaining a deeper understanding of the relationships between words and sentences and the whole text.

For example, if we come across the word "bank", we may think it means (money) bank if it is in the context of finance, but if it is in the context of fishing, etc., we may think it means river bank.

## Basic LSA application

For example, we have 4 sentences:

1. The boy who is coding.
2. The lazy yellow cat. 
3. The cat knows coding.
4. The boy is lazy.

In these four sentences, the following words appear, ordered them from A to Z, meanwhile count their frequency：
|                        | boy | cat | coding | is | knows | lazy | the | who | yellow |
|:---------------------|-----|-----|--------|----|-------|------|-----|-----|--------|
| The boy who is coding. | 1 | 0 | 1 | 1 | 0 | 0 | 1 | 1 | 0 |
| The lazy yellow cat.   | 0 | 1 | 0 | 0 | 0 | 1 | 1 | 0 | 1 |
| The cat knows coding.  | 0 | 1 | 1 | 0 | 1 | 0 | 1 | 0 | 0 |
| The boy is lazy.       | 1 | 0 | 0 | 1 | 0 | 1 | 1 | 0 | 0 |





Step 1: for the `Latent Semantic Analysis`, we want to transform these 4 sentences into a Document-Term Matrix. We are going to use the library `CountVectorizer` in Scikit-Learn.


```python
body = [
    "The boy who is coding.",
    "The lazy yellow cat.",
    "The cat knows coding.",
    "The boy is lazy.",
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(body)

display(bag_of_words.todense())
```

Step 2: applying the TSVD

```python
from sklearn.decomposition import TruncatedSVD

tsvd = TruncatedSVD(n_components=2)
lsa = tsvd.fit_transform(bag_of_words)

display(lsa)
```

<img src="data/5.png">

After we perform a full vectorizer of a text and then apply TSVD, we can get： 
- a final LSA result - for each document (in our case each sentence)
  
and we also get two by-products: 
- a full dictionary, sorted from A-Z
- the score of each word in this dictionary for each topic

```python
# topic df
topic_df = pd.DataFrame(lsa, columns=["topic_1", "topic_2"])
topic_df["body"] = body
display(topic_df[["body", "topic_1", "topic_2"]])
```

```python
# dictionary df
dictionary_df = pd.DataFrame(
    tsvd.components_.T,
    index=vectorizer.get_feature_names_out(),  # From scikit-learn 0.22 we have to use get_feature_names_out() instead of get_feature_names()
    columns=["topic_1", "topic_2"],
)

dictionary_df["abs_topic_1"] = np.abs(dictionary_df["topic_1"])
dictionary_df["abs_topic_2"] = np.abs(dictionary_df["topic_2"])

dictionary_df
```

```python
display(dictionary_df.sort_values("abs_topic_1", ascending=False))
```

```python
display(dictionary_df.sort_values("abs_topic_2", ascending=False))
```

```python
# plot the dictionary in 2D
plt.figure(figsize=(5, 5))
plt.scatter(dictionary_df["abs_topic_1"], dictionary_df["abs_topic_2"])

for word, x, y in zip(
    dictionary_df.index, dictionary_df["abs_topic_1"], dictionary_df["abs_topic_2"]
):
    plt.annotate(word, (x, y))

plt.xlabel("topic_1")
plt.ylabel("topic_2")
plt.show()
```

<font color="red"> <b> We can notice that the first topic captures the keyword coding, who scores almost 0 for the second topic. In contrast the results for the second topic are not very clear, with cat and boy having relatively high scores, suggesting that the LSA is still capturing similar themes - we are both describing something.

<font color="red"> <b> Why? - Because we are given so little data and the similarity between sentences is high

## Longer text application

Now let's test the LSA in some two larger text. The first one is a fairy tale about a little girl in the magic forest. and the second one is a zombie story about the end of world is coming.


```python
# read the two txt files in to dataframes, and keep the file name as a column
with open("The Enchanted Forest.txt", "r") as f:
    enchanted_forest = f.readlines()

with open("The Last Stand.txt", "r") as f:
    last_stand = f.readlines()

# create a dataframe with the two texts
# one column names text, include the text from the book, one column named tale
# set the column tale, if the text is from the enchanted forest, set the value to True, otherwise false

df = pd.DataFrame(
    {
        "text": enchanted_forest + last_stand,
        "tale": [True] * len(enchanted_forest) + [False] * len(last_stand),
    }
)


# remove the \n from the end of each line then remove the empty lines
df["text"] = df["text"].str.strip()
df = df[df["text"] != ""]

# shuffle the dataframe
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
df
```

Now we apply the LSA in these stories. Different than previous, here we add the paramter stop_words="english".

Stop words is a list include a bunch of words very often shown in the text but meanwhile meaningless. In the context of natural language processing, these stop_words are considered as noise, which in turn affects the topic of the text we extract.

For exemple:

a / I / of/ the / on / with / from / in / to / how / where / when / my / your / ..........

- "I am going to the store to buy some groceries."  ->  "going store buy groceries."
- "She is reading a book in the park." -> "reading book park."
- "They were playing football until it started raining."  -> "playing football started raining."


```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=1, stop_words="english")
bag_of_words = vectorizer.fit_transform(df["text"])
```

```python
from sklearn.decomposition import TruncatedSVD

tsvd = TruncatedSVD(
    n_components=2, random_state=42
)  # keep 2 components because we have 2 sources
lsa = tsvd.fit_transform(bag_of_words)
lsa
```

```python
# combian with the origin df
df1 = df.copy()
df1["topic_1"] = lsa[:, 0]
df1["topic_2"] = lsa[:, 1]
df1["abs_topic_1"] = np.abs(df1["topic_1"])
df1["abs_topic_2"] = np.abs(df1["topic_2"])

df.head(10)
```

```python
# plot two scatter plots, one is stop words removed, one is not
plt.figure(figsize=(12, 6))

# remove the stop words
plt.subplot(1, 2, 1)
plt.title("Stop words removed")
sns.scatterplot(data=df1, x="abs_topic_1", y="abs_topic_2", hue="tale")


# No remove the stop words
df2 = df.copy()
vectorizer = CountVectorizer(min_df=1)
bag_of_words = vectorizer.fit_transform(df2["text"])

tsvd = TruncatedSVD(n_components=2, random_state=42)

lsa = tsvd.fit_transform(bag_of_words)

df2["topic_1"] = lsa[:, 0]
df2["topic_2"] = lsa[:, 1]
df2["abs_topic_1"] = np.abs(df2["topic_1"])
df2["abs_topic_2"] = np.abs(df2["topic_2"])

plt.subplot(1, 2, 2)
plt.title("Stop words not removed")
sns.scatterplot(data=df2, x="abs_topic_1", y="abs_topic_2", hue="tale")
```

<font color="red"> <b> We can see that the LSA algorithm has a significant improvement in recognizing the topic of the article after removing the stop words.

```python
# visualization for the top words in the topic

vectorizer = CountVectorizer(stop_words="english")
bag_of_words = vectorizer.fit_transform(df["text"])

tsvd = TruncatedSVD(n_components=2, random_state=42)
lsa = tsvd.fit_transform(bag_of_words)

dictionary_df = pd.DataFrame(
    tsvd.components_.T,
    index=vectorizer.get_feature_names_out(),  # From scikit-learn 0.22 we have to use get_feature_names_out() instead of get_feature_names()
    columns=["topic_1", "topic_2"],
)

dictionary_df["abs_topic_1"] = np.abs(dictionary_df["topic_1"])
dictionary_df["abs_topic_2"] = np.abs(dictionary_df["topic_2"])

dictionary_df
```

```python
# scatter plot for the dictionary matrix
# if the word in both enchanted_forest and last_stand, then set to blue
# if the word in only enchanted_forest, then set to green
# if the word in only last_stand, then set to red
enchanted_forest_split = [
    word for phrase in enchanted_forest for word in phrase.split()
]
last_stand_split = [word for phrase in last_stand for word in phrase.split()]

colors = []
for i, row in dictionary_df.iterrows():
    if i in enchanted_forest_split and i in last_stand_split:
        colors.append("blue")
    elif i in enchanted_forest_split:
        colors.append("green")
    else:
        colors.append("red")

plt.figure(figsize=(6, 6))
plt.title("Dictionary matrix")
sns.scatterplot(data=dictionary_df, x="abs_topic_1", y="abs_topic_2", color=colors)

for i, row in dictionary_df.iterrows():
    if row["abs_topic_1"] > 0.2 or row["abs_topic_2"] > 0.2:
        plt.text(row["abs_topic_1"] + 0.01, row["abs_topic_2"], i)
plt.show()
```

<font color="red"> <b> We further printed the position of each word in both texts on both topics, and in addition, we distinguished each word by color:
- Words in blue appeared in both texts
- The word in green appears only in the fairy tale.
- The word in red appears only in the zombie story.

We can notice that words from fairy tales are generally clustered in the lower left corner, perhaps because of the relative simplicity of sentences and more repetitive use of words in fairy tales. In the zombie story we then get more diversity.


_We notice that some of the words appear in both texts and are clustered in the lower left corner._

## Advanced LSA application - TFIDF matrix and stop words using 20 Newsgroup dataset

```python
from sklearn.datasets import fetch_20newsgroups

newsgroups_train = fetch_20newsgroups(
    subset="train",
    # categories=["sci.crypt", "sci.electronics", "sci.med", "sci.space"],
    categories=["sci.crypt", "sci.med"],
    remove=("headers", "footers", "quotes"),
)
```

```python
# shuffle the data to make sure the body is not in order
indices = np.arange(len(newsgroups_train.data))
np.random.shuffle(indices)
newsgroups_train.data = [newsgroups_train.data[i] for i in indices]
newsgroups_train.target = newsgroups_train.target[indices]
```

```python
print("Number of documents: " + str(len(newsgroups_train.data)))
```

```python
df_20newsgroups = pd.DataFrame(
    {"text": newsgroups_train.data, "target": newsgroups_train.target}
)

# set the target name
df_20newsgroups["target_name"] = df_20newsgroups["target"].apply(
    lambda x: newsgroups_train.target_names[x]
)

df_20newsgroups.head()
```

### Text preprocessing

```python
text_data = df_20newsgroups["text"]
```

Some text contais HTML labels. Using BeautifulSoup to remove them.

```python
from bs4 import BeautifulSoup

for i in range(len(text_data)):
    soup = BeautifulSoup(text_data[i], "html.parser")
    text_data[i] = soup.get_text()
```

Secondly, we should delete punctuation marks that do not make sense. However, specific symbols may also be useful for expressing emotions and so on in statements, so here I have kept the question mark and the exclamation mark.

```python
import re
import string

punctuation = set(string.punctuation) - set(["?", "!"])
punctuation_regex = "[" + re.escape("".join(punctuation)) + "]"

cleaned_texts = [re.sub(punctuation_regex, "", text) for text in text_data]
cleaned_texts = [re.sub(r"\s+", " ", text) for text in cleaned_texts]

head_text(cleaned_texts, 2)
```

Removing all the numbers because here we focus more on the text itself.

```python
cleaned_texts = [re.sub(r"\b\d+\b", "", text) for text in cleaned_texts]

head_text(cleaned_texts, 2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD

vectorizer = CountVectorizer(stop_words="english", min_df=5)
bag_of_words = vectorizer.fit_transform(cleaned_texts)

tsvd = TruncatedSVD(
    n_components=2, random_state=42
)  # here we keep 2 components just to easily visualize, since we have 4 categories, normally we should keep 3 or 4 components
lsa = tsvd.fit_transform(bag_of_words)

# if one word is shown in more than 50 documents, then set the color to red, else set to blue

dictionary_df = pd.DataFrame(
    tsvd.components_.T,
    index=vectorizer.get_feature_names_out(),  # From scikit-learn 0.22 we have to use get_feature_names_out() instead of get_feature_names()
    columns=["topic_1", "topic_2"],
)

dictionary_df["abs_topic_1"] = np.abs(dictionary_df["topic_1"])
dictionary_df["abs_topic_2"] = np.abs(dictionary_df["topic_2"])
# dictionary_df["abs_topic_3"] = np.abs(dictionary_df["topic_3"])
# dictionary_df["abs_topic_4"] = np.abs(dictionary_df["topic_4"])

doc_freq = (bag_of_words > 0).sum(axis=0)
words = vectorizer.get_feature_names_out()
doc_freq_dict = dict(zip(words, doc_freq.tolist()[0]))

dictionary_df["color"] = dictionary_df.index.map(
    lambda x: "red" if doc_freq_dict.get(x, 0) >= 50 else "blue"
)
```

This is the problem obtained by simply counting the frequency of occurrence of words.

### With TF-IDF Matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_features=20_000,
    stop_words="english",
    min_df=5,
)  # max_features is the K we were talking about
bag_of_words = vectorizer.fit_transform(cleaned_texts)
```

### Applying TSVD

```python
n_topics = 2  # number of total topics we want to extract from the whole dataset

tsvd = TruncatedSVD(n_components=n_topics, random_state=42)
lsa = tsvd.fit_transform(bag_of_words)

topic_df = pd.DataFrame(lsa, columns=[f"topic_{i}" for i in range(n_topics)])
topic_df["text"] = cleaned_texts
topic_df["target_name"] = df_20newsgroups["target_name"]

topic_df["abs_topic_1"] = np.abs(topic_df["topic_0"])
topic_df["abs_topic_2"] = np.abs(topic_df["topic_1"])


# plot the documents in 2D topic space
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("With tfidf vectorizer and stop words removed")
sns.scatterplot(data=topic_df, x="abs_topic_1", y="abs_topic_2")

plt.subplot(1, 2, 2)
plt.title("Original target")
sns.scatterplot(data=topic_df, x="abs_topic_1", y="abs_topic_2", hue="target_name")
plt.savefig("20newsgroups_TFIDF.png")
plt.show()
```

<font color="red"> <b> with the TFIDF matrix, we reduced some of the words that appeared too frequently across documents, and thus the algorithm better recognized the two topic vectors. In our case, LSA was able to recognize a specific topic about `Medical`. But in seems that just looking at the graph on the left, his performance is still not very good, and a considerable number of documents about medical care are incorrectly categorized in topic_1.

## Implementting LSA manually

#### Manual Latent Semantic Analysis (LSA) Logic
1. Step 1: Text Preprocessing and Tokenization
    Process the cleaned texts and tokenize each document into a set of words.

2. Step 2: Creating Vocabulary
    Create a set of unique vocabulary from the words present in all documents.

3. Step 3: Building the Term-Document Matrix
- If TF-IDF is not used (tfidf=False), compute the frequency of each word in each document directly.
- If TF-IDF is used (tfidf=True), calculate the Term Frequency (TF) and Inverse Document Frequency (IDF) for each word, and    then multiply them to get the TF-IDF matrix.

4. Step 4: Performing Singular Value Decomposition (SVD)
- If manual SVD is used (use_manual_svd=True), use a custom SVD function to decompose the term-document matrix.
- If manual SVD is not used (use_manual_svd=False), use standard library functions (e.g., numpy.linalg.svd) for decomposition.

5. Step 5: Conducting LSA Processing
Extract the first n_topics topics from the singular value decomposition. This involves selecting the first n_topics columns from the U matrix and the first n_topics singular values from the S matrix, and then calculating their product to obtain the LSA matrix.

6. Step 6: Returning Results
Output the shape and content of the LSA matrix and return it as the function's result.

```python
class ManualLSA:
    """
    Manual Latent Semantic Analysis (LSA) Class writen by group 5

    - Attention: This class is not optimized for speed and memory usage!!!
    - Attention: This class supports only English and French stop words!!!
    - Attention: This class suppose that the data is already cleaned!!!

    This class implements a manual version of Latent Semantic Analysis (LSA) for text data.
    It allows for the option of using either standard SVD (Singular Value Decomposition) or a manual implementation of SVD.
    The class provides functionality to preprocess text data, construct a term-document matrix with or without TF-IDF weighting,
    perform SVD, and compute the LSA matrix.

    Key functionalities include:
    - Tokenization of text data and creation of a vocabulary set.
    - Building a term-document matrix, with an option to apply TF-IDF.
    - Performing Truncated Singular Value Decomposition (tSVD), with a choice between a manual implementation and numpy's SVD function.
    - Storing the U, Sigma, and V^T matrices internally for later use.
    - Fitting the model on text data and transforming the data into its LSA representation.

    Example usage:
        lsa = ManualLSA(tfidf=True, n_topics=3, use_manual_svd=True)
        lsa.fit(cleaned_texts)  # Fit the model on the cleaned text data
        lsa_matrix = lsa.transform()  # Get the LSA representation of the data
    """

    def __init__(
        self,
        tfidf=True,
        n_topics=5,
        stop_words="english",
        max_features=1000,
        use_manual_svd=False,
        use_df_threshold=True,
        min_df_threshold=0.05,
    ):
        self.tfidf = tfidf
        self.n_topics = n_topics
        self.stop_words = stop_words
        self.max_features = max_features
        self.use_manual_svd = use_manual_svd
        self.U = None
        self.Sigma = None
        self.VT = None
        self.LSA = None
        self.use_df_threshold = use_df_threshold
        self.min_df_threshold = min_df_threshold

    def get_stop_words(self):
        # Return the stop words
        current_dir = os.getcwd()
        import json

        path = os.path.join(current_dir, "data", "stop_words.json")

        with open(path) as f:
            stop_words = json.load(f)
        if self.stop_words == "english" or self.stop_words == "English":
            self.stop_words = stop_words["english"]
        elif self.stop_words == "french" or self.stop_words == "French":
            self.stop_words = stop_words["french"]
        else:
            self.stop_words = []

    def to_lower(self):
        # Convert the texts to lower case
        self.cleaned_texts = [text.lower() for text in self.cleaned_texts]

    def tokenize(self):
        # Tokenize the cleaned texts
        self.tokenized_texts = [
            [word for word in text.split() if word.lower() not in self.stop_words]
            for text in self.cleaned_texts
        ]

    def create_vocabulary(self):
        from collections import Counter

        # Count the frequency and document frequency of each word
        total_word_freq = Counter()
        self.document_freq = Counter()

        for text in self.tokenized_texts:
            total_word_freq.update(text)
            unique_words = set(text)
            self.document_freq.update(unique_words)

        # Set thresholds
        min_freq = 2  # Minimum frequency to include a word
        min_df = (
            len(self.tokenized_texts) * self.min_df_threshold
        )  # Minimum document frequency

        # Filter vocabulary based on frequency and document frequency
        self.vocabulary = [
            word
            for word, freq in total_word_freq.items()
            if freq >= min_freq and self.document_freq[word] >= min_df
        ]

    def build_term_document_matrix(self):
        # Initialize term-document matrix
        self.term_document_matrix = np.zeros(
            (len(self.vocabulary), len(self.tokenized_texts))
        )

        # Fill the term-document matrix
        for i, text in enumerate(self.tokenized_texts):
            for word in text:
                if word in self.vocabulary:
                    idx = self.vocabulary.index(word)
                    self.term_document_matrix[idx, i] += 1

        # If TF-IDF is used, compute TF-IDF weights
        if self.tfidf:
            df = np.array([self.document_freq[word] for word in self.vocabulary])
            idf = np.log(len(self.tokenized_texts) / (df + 1))
            idf_matrix = idf[:, np.newaxis]
            self.term_document_matrix *= idf_matrix

        elif self.tfidf:
            # Compute TF-IDF
            tf = np.zeros((len(self.vocabulary), len(self.tokenized_texts)))
            df = np.zeros(len(self.vocabulary))
            for i, text in enumerate(self.tokenized_texts):
                for word in text:
                    if word in self.vocabulary:
                        tf[self.vocabulary.index(word), i] += 1
                for word in set(text):
                    if word in self.vocabulary:
                        df[self.vocabulary.index(word)] += 1
            idf = np.log(len(self.tokenized_texts) / (df + 1))
            idf_matrix = idf[:, np.newaxis]
            self.term_document_matrix = tf * idf_matrix

        else:
            # Compute TF
            tf = np.zeros((len(self.vocabulary), len(self.tokenized_texts)))
            for i, text in enumerate(self.tokenized_texts):
                for word in text:
                    if word in self.vocabulary:
                        tf[self.vocabulary.index(word), i] += 1
            self.term_document_matrix = tf

    def manual_tsvd(self, A):
        # Compute A^TA and AA^T
        AtA = np.dot(A.T, A)
        AAt = np.dot(A, A.T)

        # Compute the eigenvalues and eigenvectors of A^TA and AA^T
        eigvals_ATA, eigvecs_ATA = np.linalg.eig(AtA)
        eigvals_AAT, eigvecs_AAT = np.linalg.eig(AAt)

        # Sort the eigenvalues and eigenvectors
        idx = eigvals_ATA.argsort()[::-1]
        eigvals_ATA = eigvals_ATA[idx]
        eigvecs_ATA = eigvecs_ATA[:, idx]

        # Compute Sigma
        Sigma = np.sqrt(np.diag(eigvals_ATA))

        # Truncate Sigma, U, and V
        Sigma = Sigma[: self.n_topics, : self.n_topics]
        U = eigvecs_AAT[:, : self.n_topics]
        V = eigvecs_ATA[:, : self.n_topics]

        return U, Sigma, V.T

    def auto_svd(self):
        if not self.use_manual_svd:
            self.U, s, self.VT = np.linalg.svd(
                self.term_document_matrix.T, full_matrices=False
            )
            # Truncate U, Sigma, and VT
            self.U = self.U[:, : self.n_topics]
            self.Sigma = np.diag(s)[: self.n_topics, : self.n_topics]
            self.VT = self.VT[: self.n_topics, :]
        else:
            self.U, self.Sigma, self.VT = self.manual_tsvd(self.term_document_matrix.T)

        # Compute LSA
        self.LSA = self.VT[: self.n_topics, :]

    def fit(self, cleaned_texts):
        self.cleaned_texts = cleaned_texts
        self.to_lower()
        self.get_stop_words()
        self.tokenize()
        self.create_vocabulary()
        self.build_term_document_matrix()
        self.auto_svd()

    def fit_transform(self, cleaned_texts):
        # Fit the model and return the LSA matrix
        self.fit(cleaned_texts)
        return self.LSA

    def get_topics(self, n_words=3):
        # Return the top n_words for each topic
        topics = []
        for i, comp in enumerate(self.LSA):
            terms_comp = zip(self.vocabulary, comp)
            sorted_terms = sorted(terms_comp, key=lambda x: x[1], reverse=True)[
                :n_words
            ]
            topics.append(sorted_terms)
        return topics

    def get_top_keywords(self, n_words=10):
        # Return the top n_words keywords
        from collections import Counter

        keyword_freq = Counter()

        for i, comp in enumerate(self.LSA):
            terms_comp = zip(self.vocabulary, comp)
            sorted_terms = sorted(terms_comp, key=lambda x: x[1], reverse=True)
            for term, _ in sorted_terms:
                keyword_freq[term] += 1

        top_keywords = [word for word, freq in keyword_freq.most_common(n_words)]
        return top_keywords
```

### Parameter Tuning in LSA

`CountVectorizer Parameters`

- **max_df**:
  - Used to remove terms that appear too frequently, represented as a proportion (e.g., 0.5 for 50%) or an absolute count.
  - Helps in eliminating common words that appear in most of the documents and are less informative about the topic.

- **min_df**:
  - Used to remove terms that appear too infrequently, also can be a proportion or an absolute count.
  - Useful for excluding rare words that appear only in a few documents.

- **max_features**:
  - Limits the size of the vocabulary to the top N terms by frequency.
  - This is helpful in reducing model complexity and improving training speed.

- **stop_words**:
  - Can be set to a list of stop words for a particular language to remove words that typically contribute little to the meaning of a text (e.g., "the", "is").
  
- **ngram_range**:
  - Sets the range of n-values for different n-gram or multi-word combinations (e.g., (1, 2) includes both single words and 2-word combinations).
  - Valuable for capturing more information about word groups and context.


`TfidfVectorizer Parameters`

- **max_df**:
  - Removes words that appear in a higher proportion of documents than specified.
  - Useful for eliminating ubiquitous terms that are less informative about the topic.

- **min_df**:
  - Removes words that appear in a lower proportion of documents than specified.
  - Helps in discarding rare words that appear only in a few documents.

- **max_features**:
  - Restricts the vocabulary size to the top N terms based on term frequency.
  - Aids in reducing model complexity and training time.

- **stop_words**:
  - Can specify a list of stop words to be removed from the text.
  - Useful for eliminating common words that carry little meaningful information.

- **ngram_range**:
  - Determines the range of n-values for different n-gram combinations.
  - Important for capturing information about word sequences and context.

- **use_idf**:
  - Whether to use inverse document frequency (IDF) weighting; default is True.
  - Weighs down terms that appear more frequently across documents and gives more importance to rarer terms.

- **smooth_idf**:
  - Smoothes IDF weights by adding 1 to document frequencies, as if an extra document contained every term in the collection exactly once. Prevents divide-by-zero errors.
  - Provides a gentler IDF weighting for terms, particularly useful in smaller corpora.

- **norm**:
  - Norm used to normalize term vectors; either 'l2' (default) for Euclidean or 'l1' for Manhattan norm.
  - Ensures all feature vectors have a consistent length, beneficial for downstream models and computations.


### Comparing LSA with Other Methods

<font size='5.5'><b> Better solutions... </font></b><br>

LSA -> PLSA -> LDA -> BERT <br><br>

* Probabilistic Latent Semantic Analysis (pLSA) improves upon LSA by introducing a probabilistic model. It models each word in a document as a sample from a mixture model, where the mixture components are multinomial random variables that can be thought of as "topics". pLSA provides a probabilistic foundation which LSA lacks, making it easier to handle uncertainties and variabilities in word distributions. However, pLSA has the tendency to overfit, especially with a large number of parameters to estimate when the corpus size increases. <br>

* Latent Dirichlet Allocation (LDA) extends pLSA by placing Dirichlet priors on the distributions over topics in documents and words in topics. This Bayesian approach helps in handling overfitting and makes LDA more suitable for dealing with larger corpora. The Dirichlet priors act as a regularization mechanism, allowing the model to generalize better to unseen documents. LDA can be more interpretable and flexible, and it provides a full generative model of the corpus, which means that it can be used to generate new documents that resemble the original ones.

* Bidirectional Encoder Representations from Transformers (BERT) outshines LSA by leveraging a transformer-based architecture for in-depth contextual word analysis. It surpasses LSA's singular value decomposition by considering bidirectional context, offering a more nuanced understanding of language. BERT's pre-training on extensive text data allows it to perform exceptionally well on diverse NLP tasks, including topic modeling, sentiment analysis, question answering, and named entity recognition, by applying its sophisticated language model, which excels in handling polysemy (words with multiple meanings) and complex sentences.
